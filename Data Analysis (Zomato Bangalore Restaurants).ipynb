```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("zomato.csv")
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df=df.drop(['url', 'address','phone','dish_liked','reviews_list'],axis=1)
df.head(10)
```

```python
df=df.drop(['menu_item'],axis=1)
df.head(10)
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def rate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(rate)
df['rate'].head(10)
```

```python
df.isnull().sum()
```

```python
# feature engineering
df['rate'].fillna(df['rate'].mean(), inplace=True)
```

```python
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head(10)
```

```python
df.rename(columns={'approx_cost(for two people)':'2-PlatesCost','listed_in(type)':'Type','listed_in(city)':'City'},inplace=True)
df.head(10)
```

```python
df.rename(columns={'name':'Name','online_order':'Online Order','book_table':'Book Table','rate':'Rating','votes':'Votes','location':'Location','rest_type':'Restaurant Type','cuisines':'Cuisines'},inplace=True)
df.head()
```

```python
df['Location'].unique()
```

```python
df['City'].unique()
```

```python
df.drop(['City'],axis=1)
```

```python
df['2-PlatesCost'].unique()
```

```python
def plates_cost(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['2-PlatesCost']=df['2-PlatesCost'].apply(plates_cost)
df['2-PlatesCost'].head(10)
```

```python
df['Restaurant Type'].value_counts()
```

```python
Res_type=df['Restaurant Type'].value_counts(ascending=False)
Res_type
```

```python
Res_type_lessthan1k=Res_type[Res_type<1000]
Res_type_lessthan1k
```

```python
def Res_type(value):
    if(value in Res_type_lessthan1k):
        return 'Others'
    else:
        return value
df['Restaurant Type']=df['Restaurant Type'].apply(Res_type)
df['Restaurant Type'].value_counts()
```

```python
df.head(10)
```

```python
df['Location'].value_counts()
```

```python
Location=df['Location'].value_counts(ascending=False)
```

```python
Location_lessthan500=Location[Location<500]
```

```python
def location(value):
    if(value in Location_lessthan500):
        return 'Others'
    else:
        return value
df['Location']=df['Location'].apply(location)
df['Location'].value_counts()
```

```python
df.head(10)
```

```python
df['Cuisines'].value_counts()
```

```python
Cuisines=df['Cuisines'].value_counts(ascending=False)
```

```python
Cuisines_lessthan100=Cuisines[Cuisines<100]
```

```python
def cuisines(value):
    if(value in Cuisines_lessthan100):
        return 'Others'
    else:
        return value
df['Cuisines']=df['Cuisines'].apply(cuisines)
df['Cuisines'].value_counts()
```
